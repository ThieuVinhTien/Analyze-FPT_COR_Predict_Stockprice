```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Read the CSV file
df = pd.read_csv('D:/Desktop/Profit -- Stock price.csv')
```

```python
df.set_index('DATE', inplace=True)
```

```python
df
```

```python
# Add a constant term for intercept
x = df['CLOSE']
X = sm.add_constant(x)

Y = df['PROFITAFTERTAX']
```

```python
np.asarray(df)
```

```python
X
```

```python
Y
```

```python
# Fit the multiple regression model
model = sm.OLS(Y, X).fit()

# Print the summary of the regression results
print(model.summary())
```
